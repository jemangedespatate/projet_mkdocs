<h1><span class="title-number">III</span>Phase 2 : L'algorithme k-NN optimisé</h1>

Nous allons réutiliser tes connaissances mais avec une structure de code plus professionnelle.

```python
def distance(f1, f2):
    # f1['caract'] est une liste de 4 nombres
    d_carre = 0
    for i in range(4):
        d_carre += (f1['caract'][i] - f2['caract'][i])**2
    return math.sqrt(d_carre)

def predire_espece(entrainement, cible, k):
    # 1. Calculer les distances
    distances = []
    for fleur in entrainement:
        d = distance(fleur, cible)
        distances.append((d, fleur['espece']))
    
    # 2. Trier
    distances.sort()
    
    # 3. Prendre les k premiers voisins
    voisins = distances[:k]
    
    # 4. Voter (majorité)
    compteurs = {}
    for d, espece in voisins:
        compteurs[espece] = compteurs.get(espece, 0) + 1
    
    # Retourner l'espèce majoritaire
    max_votes = 0
    vainqueur = ""
    for espece, votes in compteurs.items():
        if votes > max_votes:
            max_votes = votes
            vainqueur = espece
    return vainqueur
```

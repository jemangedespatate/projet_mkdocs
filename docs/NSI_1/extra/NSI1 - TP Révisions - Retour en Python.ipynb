# 🔴 Partie 4 — Algorithme de tri

## Le tri par sélection

Principe :
1. On cherche le **plus petit élément** → on l'échange avec la position 0.
2. On cherche le plus petit dans le reste → on l'échange avec la position 1.
3. On répète jusqu'à avoir tout trié.

```
[7, 3, 5, 2]  →  min=2, échange pos 0↔3  →  [2, 3, 5, 7]
[2, 3, 5, 7]  →  min=3, déjà en place    →  [2, 3, 5, 7]  ✅
```

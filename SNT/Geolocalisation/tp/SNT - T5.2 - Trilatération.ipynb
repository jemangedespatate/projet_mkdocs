<h1><span class="title-number">V</span>Sur une carte</h1>

Le code Python suivant affiche une **carte interactive** de la Picardie, ainsi que la position du pylône d’_Aumale_ :

```python
# ============================================================
# ETAPE 1 : Importer la bibliotheque folium
# ------------------------------------------------------------
# 'folium' est une bibliotheque Python qui permet de creer des
# cartes interactives a partir de donnees geographiques.
# On l'importe une seule fois au debut du programme.
# ============================================================
import folium

# ============================================================
# ETAPE 2 : Creer la carte de base
# ------------------------------------------------------------
# folium.Map() cree un objet 'carte'.
# Parametres :
#   - location=[latitude, longitude] : coordonnees GPS du centre
#     de la carte. Ici, on centre sur la Picardie.
#     latitude  = 49.6531  (nord-sud : + = Nord, - = Sud)
#     longitude = 2.5841   (est-ouest : + = Est, - = Ouest)
#   - zoom_start=8 : niveau de zoom initial (1=monde, 18=rue)
# La carte est stockee dans la variable 'maCarte'.
# ============================================================
maCarte = folium.Map(location=[49.6531, 2.5841], zoom_start=8)

# ============================================================
# ETAPE 3 : Ajouter un marqueur sur le pylone d'Aumale
# ------------------------------------------------------------
# folium.Marker() cree un marqueur (epingle) sur la carte.
# Parametres :
#   - location=[lat, lon] : position GPS du marqueur
#     Ici, le pylone d'Aumale est a :
#     latitude  = 49.7719 degres Nord
#     longitude = 1.7538  degres Est
#   - tooltip='Aumale' : texte affiche au survol de la souris
#   - icon=folium.Icon(...) : personnalise l'icone du marqueur
#       color='red'       -> couleur rouge du marqueur
#       icon='info-sign'  -> petit 'i' d'information
# .add_to(maCarte) : ajoute le marqueur a notre carte.
# ============================================================
folium.Marker(
    location=[49.7719, 1.7538],
    tooltip="Aumale",
    icon=folium.Icon(color='red', icon='info-sign')
).add_to(maCarte)

# ============================================================
# ETAPE 4 : Afficher la carte
# ------------------------------------------------------------
# En ecrivant simplement 'maCarte' a la fin d'une cellule
# Jupyter, la carte interactive s'affiche dans le notebook.
# On peut zoomer, se deplacer et cliquer sur les marqueurs !
# ============================================================
maCarte
```

````python
# ============================================================
# ETAPE 1 : Importer la bibliotheque folium
# ------------------------------------------------------------
# 'folium' est une bibliotheque Python qui permet de creer des
# cartes interactives a partir de donnees geographiques.
# On l'importe une seule fois au debut du programme.
# ============================================================
import folium

# ============================================================
# ETAPE 2 : Creer la carte de base
# ------------------------------------------------------------
# folium.Map() cree un objet 'carte'.
# Parametres :
#   - location=[latitude, longitude] : coordonnees GPS du centre
#     de la carte. Ici, on centre sur la Picardie.
#     latitude  = 49.6531  (nord-sud : + = Nord, - = Sud)
#     longitude = 2.5841   (est-ouest : + = Est, - = Ouest)
#   - zoom_start=8 : niveau de zoom initial (1=monde, 18=rue)
# La carte est stockee dans la variable 'maCarte'.
# ============================================================
maCarte = folium.Map(location=[49.6531, 2.5841], zoom_start=8)

# ============================================================
# ETAPE 3 : Ajouter un marqueur sur le pylone d'Aumale
# ------------------------------------------------------------
# folium.Marker() cree un marqueur (epingle) sur la carte.
# Parametres :
#   - location=[lat, lon] : position GPS du marqueur
#     Ici, le pylone d'Aumale est a :
#     latitude  = 49.7719 degres Nord
#     longitude = 1.7538  degres Est
#   - tooltip='Aumale' : texte affiche au survol de la souris
#   - icon=folium.Icon(...) : personnalise l'icone du marqueur
#       color='red'       -> couleur rouge du marqueur
#       icon='info-sign'  -> petit 'i' d'information
# .add_to(maCarte) : ajoute le marqueur a notre carte.
# ============================================================
folium.Marker(
    location=[49.7719, 1.7538],
    tooltip="Aumale",
    icon=folium.Icon(color='red', icon='info-sign')
).add_to(maCarte)

# ============================================================
# ETAPE 4 : Afficher la carte
# ------------------------------------------------------------
# En ecrivant simplement 'maCarte' a la fin d'une cellule
# Jupyter, la carte interactive s'affiche dans le notebook.
# On peut zoomer, se deplacer et cliquer sur les marqueurs !
# ============================================================
maCarte
````

La ligne de code suivante permet d’ajouter un **cercle** centré sur Aumale de rayon 61 km :

```python
# ============================================================
# Ajouter un cercle autour du pylone d'Aumale
# ------------------------------------------------------------
# folium.Circle() dessine un cercle sur la carte.
# Parametres :
#   - location=[lat, lon] : centre du cercle (position du pylone)
#   - radius=61000 : rayon en METRES (61000 m = 61 km)
#     -> Cette valeur correspond a la distance calculee
#        entre le pylone et le tresor (Question 4).
#     -> Tous les points situes sur ce cercle sont a 61 km du pylone.
#     -> Le tresor se trouve quelque part sur ce cercle !
#
# Parametres optionnels utiles :
#   - color='blue'      : couleur du bord du cercle
#   - fill=True         : remplir l'interieur du cercle
#   - fill_opacity=0.1  : transparence du remplissage (0 a 1)
#   - tooltip='...'     : texte affiche au survol de la souris
#
# .add_to(maCarte) : ajoute le cercle a notre carte.
# ============================================================
folium.Circle(location=[49.7719, 1.7538], radius=61000).add_to(maCarte)
```
